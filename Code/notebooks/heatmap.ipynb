```python
def combine(array):
    combinedImage = array[0]
    for i in range(len(array)):
        for j in range(100):
            for k in range(100):
                for rgb in range(3):
                    if combinedImage[j][k][rgb] < array[i][j][k][rgb]:
                        combinedImage[j][k][rgb] = array[i][j][k][rgb]


    return combinedImage
```
